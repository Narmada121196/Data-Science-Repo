```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read the dataset
df = sns.load_dataset('tips')
df.head() 
```

#### 1.Create a Line Plot with x (‘total_bill’) versus y (‘tip’) using lineplot() function:

```python
sns.lineplot(x = 'total_bill',y = 'tip', data = df)
```

###### Explanation

```python
Here the lineplot says that, total bill is directly propotional to the tip, which means the tip increases
when the total_bill increases and vice versa.
```

```python
sns.lineplot(x = 'total_bill',y = 'tip', data = df, hue = 'day',palette = 'gist_heat')
```

###### Explanation
This is a lineplot of total_bill vs tip based on day, this clearly says tip is given the highest on
"Saturday". But, all the days there is a tip between 2 to 4.


```python
sns.lineplot(x = 'total_bill',y = 'tip', data = df, hue = 'time' , palette = 'icefire')
```

###### Explanation
This is a lineplot of total_bill vs tip based on "Time", it clearly says during "dinner" the tips 
are high.

```python
sns.lineplot(x = 'total_bill',y = 'tip', data = df, hue = 'sex' , palette = 'icefire')
```

###### Explanation
This is a lineplot of total_bill vs tip based on "Sex", which implies "Male" tips more than the female.

```python
sns.lineplot(x = 'total_bill',y = 'tip', data = df, hue = 'size' , palette = 'mako')
```

```python
The relationship between total_bill vs tip based on "size" is not correlated. 
It does not imply the higher bill or higher sized gives more tip or the lower bill or lower sized 
gives lower tip.
```

## 4.Write a python code to create a heatmap on tips dataset by showing the corelation between numeric attributes 

```python
sns.heatmap(df.corr())
```

###### Explanation
By side of the heatmap, the colour shades are given with the values which says 1.0 are highly correlated and 0.5 is where the correlations are low and the points between are moderately correlated.

total_bill vs Size::
This gives a value between 0.6 and below 0.7 which says they are correlated to some extend and not
higher or lower correlation.

total_bill vs tip::
This gives a value higher than the comparison between total_bill vs size ranging from 0.68 to 0.75
which means it is highly correlated than the total_bill vs size. In general, it is moderately correlated.

tip vs size::
This gives a value very lower than the comparison between total_bill vs size and 
total_bill vs tip as ranging between 0.5 to 0.578,which means it is not correlated at all.

So, the correlation hierarchy is:

total_bill vs tip > total_bill vs Size > tip vs size

Non-Smoker - Male vs Female  :: "Male" pays the higher bill (21.5) than "Female" (18)
Male - smoker vs non-smoker  :: "Male" who smokes pays higher bill (22.5 or 23) than the male who do                                  not smoke(21.5).

Female - smoker vs non-smoker:: "Female" who smoke (17.5) pays equal or little less than the 
                                 non-smoking "Female"(18).
                                 
Male vs Female - smoker vs non-smoker :: Considering all factors together, it is evident that Male                                            pay higher bill than the Female.
