# Question 4
If we look at the data on a weekly basis, which country rapidly increased their testing rate the most.

````python
df.head()
````

First thing first, the way we have our week column setup right now 
- it indicates the week number(out of 52) of the year, 
- when the next year starts, it again starts from 0.

What if we don't want this behaviour, and we want the next year 1st week to be the '53th' week instead.
- How do we do this
- Well, we can just add 52 to the next year weeks, and 104 to the year next to it and so on.

So, for how many years do we have the data for. Let's find out

````python
df['date'].dt.isocalendar().year.unique()
````

We have data for 3 years, so simple enough
- i'll add 52 weeks to the rows with the year 2021, and 104 to the rows with year 2022. 

````python
df['Year'] = df['date'].dt.isocalendar().year
````

````python
# Let's not touch 2020
df[df['Year'] == 2020]
````

````python
# This is our 2021 data
df[df['Year'] == 2021]
````

````python
# Let's add 52 to our 2021 data 
df[df['Year'] == 2021]['week_of_the_year'] + 52
````

Now, we will need to save this in our dataframe.
- How do we do this. 
- Well, we can overwrite the rows

````python
# So you may think, this would work 
df[df['Year'] == 2021]['week_of_the_year'] = df[df['Year'] == 2021]['week_of_the_year'] + 52
````

````python
# It didn't work, values still the same
df[df['Year'] == 2021]['week_of_the_year']
````

**But it won't**. Look [here.](https://www.dataquest.io/blog/settingwithcopywarning/) for more knowledge about ```SettingWithCopyWarning```.

- So, i have to use iloc to do this.
- i am using the index of ```df[df['Year'] == 2021]['week_of_the_year']``` to call my rows with iloc and using ```-2``` to get the 2nd last column, which is our week column.

````python
# Getting our desired rows and columns with iloc
df.iloc[df[df['Year'] == 2021]['week_of_the_year'].index,-2]
````

````python
df.iloc[df[df['Year'] == 2021]['week_of_the_year'].index,-2] = df[df['Year'] == 2021]['week_of_the_year'] + 52
````

# So, for those reason, this code works
![](https://media.tenor.com/8pLlGDVzPRYAAAAM/muppet-family-christmas-muppets.gif)

````python
# IT WORKED !!!!
df[df['Year'] == 2021]['week_of_the_year']
````

### Alright, let's do the same for 2022

````python
df[df['Year'] == 2022]['week_of_the_year']
````

Yup, looks like the same issue to me.
- So doing the same thing
- But adding 104 this time around

````python
df[df['Year'] == 2022]['week_of_the_year'] + 104
````

Now, to save it 

````python
# Using the same logic as above
df.iloc[df[df['Year'] == 2022]['week_of_the_year'].index,-2] = df[df['Year'] == 2022]['week_of_the_year'] + 104
````

````python
# It worked !!
df[df['Year'] == 2022]['week_of_the_year']
````

# Coming back to the question
If we look at the data on a weekly basis, which country rapidly increased their testing rate the most.

Let's figure out the **testing rate** part

````python
df.head()
````

````python
df.columns
````

First of all, to get the testing rate, we will first need to get the percent change, so if today we had 100 tests and tommorow we have 102. We have a 2% change (102-100)/100. We need to calculate this for every country

````python
# Getting the rows for india
df[df['location'] == 'India']
````

````python
# Grouping by the week
df[df['location'] == 'India'].groupby('week_of_the_year').sum()
````

````python
# Getting only the new test column
df[df['location'] == 'India'].groupby('week_of_the_year').sum()['new_tests']
````

# Now focus here
![](https://media.tenor.com/ze3REKFql9YAAAAC/cat-focus.gif)

We need to calculate the change rate. One way is to do this manually
- Like this

````python
# I am not going to do this manually
# (df['new_tests'] - df['new_tests'].shift(1))/df['new_tests']
````

What you can use is Pandas DataFrame ```pct_change()``` Method
- The ```pct_change()``` method returns a DataFrame with the percentage difference between the values for each row and, by default, the previous row.

````python
df[df['location'] == 'India'].groupby('week_of_the_year').sum()['new_tests'].pct_change()
````

Let's save this

````python
rate_df = df[df['location'] == 'India'].groupby('week_of_the_year').sum()['new_tests'].pct_change() 
````

````python
rate_df
````

### Now, for the plotting

````python
rate_df.plot()
````

# So, for india, this is what we have. 
- We see a rapid increase in testing rates in the beginning, and then a tappering off, after 30 weeks, we found an equilibrium and didn't see a lot of increase or decrease in testing, it remained overall the same.

````python
# Let's fill the NA's with 0
df.groupby(['location','week_of_the_year']).sum()['new_cases'].fillna(0).pct_change()
````
